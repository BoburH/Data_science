# Numpy kutibxonasining universal funksiyalari

```python
import numpy as np
```

# Unary funksiyalar. bunday funksiyalar faqat bitta argument qabul qiladi.


*   sqrt, square, exp, log, modf, sign, isnan funksiyalarini ko'rib chiqamiz



```python
arr = np.arange(10)
arr
```

```python
np.sqrt(arr)     # massivning har bir elementi kvadrat ildizi
```

```python
np.square(arr)    # massivning har bir elementi kvadrati
```

```python
np.exp(arr)      # e ning massiv elementlari darajasida
```

```python
np.log(arr[1:])       # massiv elementlari logorifmlari
```

```python
arr1= np.random.randn(6)      # normal distributionli tasodifiy massiv
arr1
```

```python
np.modf(arr1)      # arr1 ning elementlarini aloxida massivlar orqali butun va kasr qismiga bo'lib beradi.
```

```python
np.isnan(arr1)     # bu massivga so'rov berish uchun ya'ni qaysi elementlar ma'lumot yo'q degani
```

```python
arr1[5] = np.NaN      # arr1 massivining 5 indeksli elementini NaNga tenglab olmoqdamiz
```

```python
np.isnan(arr1)     # True qaytargan element ma'lumot mavjud emas degani
```

```python
np.sign(arr1)     # sign orqali elementlar manfiy yoki musbatligi tekshiriladi -1 qaytargani manfiy, 1 qaytargani musbat bo'ladi.
```

# Binary funksiyalar, bunday funksiyalar 2 ta argument qabul qiladi
# add, multiply, maximum funksiyalrini ko'rib chiqamiz

```python
arr2 = np.random.randint(5, 15, 6)
arr2
```

```python
arr3 = np.random.randint(1, 10, 6)
arr3
```

```python
# add qo'shuv amali
np.add(arr2, arr3)
```

```python
arr2 + arr3
```

```python
# multiply ko'paytiruv amali
np.multiply(arr2, arr3)
```

```python
arr2 * arr3
```

```python
# maximum amali massivlarning elementlarini solishtirib kattasini aniqlaydi
print(arr2)
print(arr3)
```

```python
np.maximum(arr2, arr3)
```
