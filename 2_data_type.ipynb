# Ma'lumotlar tahlili

```python
import numpy as np
```

image.png

```python
arr1 = np.array([1, 2, 3], dtype=np.float64)    # bu float64 ma'lumotlar turiga ega massiv
```

```python
arr1
```

```python
arr2 = np.array([1, 2, 3], dtype=np.int64)     # bu int64 ma'lumotlar turiga ega massiv
```

```python
arr2
```

```python
arr1.dtype   # .dtype metodi massiv ma'lumotlari qaysi turda ekanligini ko'rsatadi
```

# Bir ma'lumot turidan boshqa ma'lumot turiga o'zgartirish

```python
arr3 = np.array([1, 2, 3, 4])     # Bu default rafishda int64 turida massiv yaratadi
```

```python
arr3
```

```python
arr3.dtype
```

```python
float_arr = arr3.astype(np.float64)     # astype metodi orqali o'zgartiriladi
```

```python
float_arr
```

```python
float_arr.dtype
```

```python
data = np.array([3.2, 4.2, 5.5, 6.4, 7.8])     # bu default rafishda float64 da yaratiladi
```

```python
data.dtype
```

```python
int_data = data.astype(np.int64)
```

```python
int_data
```

```python
int_data.dtype
```
