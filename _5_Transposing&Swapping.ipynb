# Massivlar positsiyasini ko'chirish transpose va almashtirish swapping

```python
import numpy as np
```

image.png

```python
# reshape metodi

A = np.arange(6).reshape(3,2)    # reshape metodi yordamida bir o'lchamli yaratilgan massivni ikki o'lchamli holatga o'tkazishimiz mumkin.
A
```

```python
# Transpose
A.T    # transpose ko'rinishini berish uchun .T ni qo'llash kifoya
```

```python
# Swaping axes ya'ni yo'nalishini o'zgartirish uchun
A
```

```python
A.swapaxes(1,0)
```

```python
np.dot(A, A.T)      # dot yordamida ikkita massivni ko'paytiramiz
```
